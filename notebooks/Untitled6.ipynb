Need averages from USGS Statistics report

```python

url = "https://waterservices.usgs.gov/nwis/stat/?format=rdb&sites={sites}&parameterCd={parameter}&statReportType=daily"
```

```python
sites = ','.join(gages.STAID.values[:10])
```

```python
url.format(sites = sites, parameter='00060')
```

```python
r = requests.get(url.format(sites = sites, parameter='00060'))
```

```python
allsites = pd.DataFrame()
for start in range(0, len(gages), 10):
    sites = ','.join(gages.STAID.values[start:start+10:])
    allsites = pd.concat([allsites, pd.read_csv(url.format(sites = sites, parameter='00060'), sep='\t', comment='#').drop(0, axis=0)])
```
